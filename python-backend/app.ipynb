```python
import yfinance as yf

def get_percentage_change(symbol):
    stock = yf.Ticker(symbol)

    periods = {
        "1d": "1d",
        "1w": "5d",
        "1m": "1mo",
        "3m": "3mo",
        "6m": "6mo",
        "1y": "1y",
        "5y": "5y"
    }

    results = {}

    for key, period in periods.items():
        try:
            hist = stock.history(period=period)
            if len(hist) > 1:
                start_price = hist["Close"].iloc[0]
                end_price = hist["Close"].iloc[-1]
                percent_change = ((end_price - start_price) / start_price) * 100
                results[key] = round(percent_change, 2)
            else:
                results[key] = None
        except Exception as e:
            results[key] = None

    return results


if __name__ == "__main__":
    symbol = input("Enter stock symbol (e.g., TCS.NS, AAPL): ").upper()
    changes = get_percentage_change(symbol + ".NS")

    print(f"\nPercentage Change for {symbol}:")
    for period, value in changes.items():
        if value is not None:
            print(f"{period.upper()}: {value}%")
        else:
            print(f"{period.upper()}: No data available")
```

```python
import yfinance as yf


def normalize_symbol(symbol: str, default_suffix: str = ".NS") -> str:
    """Normalize the ticker symbol.

    If the symbol does not include a market suffix (e.g., ".NS"), this will append the default.
    """
    symbol = symbol.strip().upper()
    if "." not in symbol:
        symbol = f"{symbol}{default_suffix}"
    return symbol


def get_percentage_change(symbol: str) -> dict:
    stock = yf.Ticker(symbol)

    periods = {
        "1d": "1d",
        "1w": "5d",
        "1m": "1mo",
        "3m": "3mo",
        "6m": "6mo",
        "1y": "1y",
        "5y": "5y"
    }

    results = {}

    for key, period in periods.items():
        try:
            hist = stock.history(period=period)
            if len(hist) > 1:
                start_price = hist["Close"].iloc[0]
                end_price = hist["Close"].iloc[-1]
                percent_change = ((end_price - start_price) / start_price) * 100
                results[key] = round(percent_change, 2)
            else:
                results[key] = None
        except Exception:
            results[key] = None

    return results


if __name__ == "__main__":
    # Provide a list of stocks to calculate percentage changes for.
    # You can add or remove symbols as needed.
    stocks = ["TCS", "RELIANCE", "INFY"]

    for stock in stocks:
        symbol = normalize_symbol(stock)
        changes = get_percentage_change(symbol)

        print(f"\nPercentage Change for {symbol}:")
        for period, value in changes.items():
            if value is not None:
                print(f"{period.upper()}: {value}%")
            else:
                print(f"{period.upper()}: No data available")
```
